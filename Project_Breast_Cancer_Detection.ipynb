#Breast Cancer Detection

##Importing the Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##Data collection and Preprocessing

```python
from sklearn.datasets import load_breast_cancer as breast_cancer_dataset
data = breast_cancer_dataset()
```

```python
print(data)
```

```python
# loading the dataframe into pandas for better view
df = pd.DataFrame(data.data , columns= data.feature_names)
```

```python
df.head()
```

```python
# adding the target column
df['label'] = data.target
```

```python
df.tail()
```

```python
# printing the number of rows and columns in the dataframe
df.shape
```

```python
#getting the information about the data
df.info()
```

```python
# method to check for missing values.
df.isnull().sum()
```

```python
# statistical measures about the data
df.describe()
```

```python
#checking the distribution of the target variable
df['label'].value_counts()
```

### here in the above checkings
0 -> Benign Cancer Type (More dangerous)

---


1 -> Malignant Cancer Type

```python
df.groupby('label').mean()
```

#Splitting the data by using the sci-kit learn library

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

#Splitting the data into training data and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Feature Scailing

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#Model Training

#Logistic Regression

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

#Evaluating the model

###Accuracy Score


```python
X_train_pred = model.predict(X_train)
```

```python
#comparing the y_train and X_train
training_data_accuracy = accuracy_score(y_train, X_train_pred)
```

```python
print(training_data_accuracy)
```

```python
X_test_pred = model.predict(X_test)
```

```python
training_data_accuracy_test = accuracy_score(y_test, X_test_pred)
```

```python
print(training_data_accuracy_test)
```

#Building a predictive system

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# Change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)
# reshaping the model for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
if(prediction[0] == 0):
  print('The Breast cancer is Malignant')

else:
  print('The Breast Cancer is Benign')
```

#Visualization

```python
# Create bar plot
plt.figure(figsize=(8, 6))
plt.bar(['Training Accuracy', 'Test Accuracy'], [training_data_accuracy, training_data_accuracy_test], color=['blue', 'green'])
plt.ylim([0, 1])  # Set y-axis limits from 0 to 1
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')

# Add text labels on top of each bar
for i, v in enumerate([training_data_accuracy, training_data_accuracy_test]):
    plt.text(i, v + 0.01, f'{v:.4f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
importances = model.coef_[0]

# Sort features by importance
indices = np.argsort(np.abs(importances))
top_k = 10  # Show top 10 features

plt.figure(figsize=(10, 6))
plt.title(f"Top {top_k} Feature Importances")
plt.bar(range(top_k), np.abs(importances[indices[-top_k:]]))
plt.xticks(range(top_k), [f"Feature {i}" for i in indices[-top_k:]], rotation=45, ha='right')
plt.xlabel("Features")
plt.ylabel("Absolute Importance")
plt.tight_layout()
plt.show()
```
